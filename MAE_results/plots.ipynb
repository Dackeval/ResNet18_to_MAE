```python
import matplotlib.pyplot as plt
import json

def plot_results(file1, file2, file3):
    # 读取文件并解析 JSON
    with open(file1, 'r') as f:
        data1 = json.load(f)
        train_loss_1 = data1['loss']
    
    with open(file2, 'r') as f:
        data2 = json.load(f)
        train_loss_2 = data2['train_loss']
        val_loss_2 = data2['val_loss']
        acc_train_2 = data2['acc_train']
        acc_val_2 = data2['acc_val']
    
    with open(file3, 'r') as f:
        data3 = json.load(f)
        train_loss_3 = data3['train_loss']
        val_loss_3 = data3['val_loss']
        acc_train_3 = data3['acc_train']
        acc_val_3 = data3['acc_val']
    
    # 创建一个图形框，在其中可以放置多个图
    fig, axs = plt.subplots(1, 3, figsize=(15, 5))
    
    # 第一个图：Pretrain的train_loss曲线
    axs[0].plot(train_loss_1, label='Pretrain Loss', color='blue')
    axs[0].set_title('Pretrain Train Loss')
    axs[0].set_xlabel('Epoch')
    axs[0].set_ylabel('Loss')
    axs[0].legend()
    
    # 第二个图：Linear Probing和Fine Tuning的accuracy曲线
    axs[1].plot(acc_train_2, label='From Scratch Train Acc', color='red')
    axs[1].plot(acc_val_2, label='From Scratch Val Acc', color='blue')
    # axs[1].plot(acc_train_3, label='Fine Tuning Train Acc', color='green')
    # axs[1].plot(acc_val_3, label='Fine Tuning Val Acc', color='black')
    axs[1].set_title('Accuracy Comparison')
    axs[1].set_xlabel('Epoch')
    axs[1].set_ylabel('Accuracy')
    axs[1].legend()
    
    # 第三个图：Linear Probing和Fine Tuning的loss曲线
    axs[2].plot(train_loss_2, label='From Scratch Train Loss', color='red')
    axs[2].plot(val_loss_2, label='From Scratch Val Loss', color='blue')
    # axs[2].plot(train_loss_3, label='Fine Tuning Train Loss', color='green')
    # axs[2].plot(val_loss_3, label='Fine Tuning Val Loss', color='black')
    axs[2].set_title('Loss Comparison')
    axs[2].set_xlabel('Epoch')
    axs[2].set_ylabel('Loss')
    axs[2].legend()
    
    # 显示图
    plt.tight_layout()
    plt.show()
```

```python
plot_results('pretrain_mask_ratio_0.75_decoder_depth_8', 'from_scratch_mae_pretrain_maskratio_0.75_dec_depth_4', 'fine_tune_maskratio_0.75_dec_depth_8')
```
