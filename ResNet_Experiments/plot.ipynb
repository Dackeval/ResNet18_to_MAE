```python
import json
import matplotlib.pyplot as plt
```

```python
with open('/Users/tobiaspeihengli/Downloads/DD2424/Project/DD2424-final-project/Results/da_results', 'r') as file:
    data = json.load(file)
```

```python
# Plotting variables
epochs = list(range(1, 11))  # Assuming there are 10 epochs

# Create figure for Loss
plt.figure(figsize=(10, 6))
for i, aug_type in enumerate(data['augmentation_type']):
    train_loss = data['final_acc'][i]['train_loss']
    val_loss = data['final_acc'][i]['val_loss']
    plt.plot(epochs, train_loss, marker='o', label=f'Train Loss ({aug_type})')
    plt.plot(epochs, val_loss, marker='x', linestyle='--', label=f'Val Loss ({aug_type})')
plt.title('Training and Validation Loss Across Augmentations')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Create figure for Accuracy
plt.figure(figsize=(10, 6))
for i, aug_type in enumerate(data['augmentation_type']):
    train_acc = data['final_acc'][i]['train_acc']
    val_acc = data['final_acc'][i]['val_acc']
    plt.plot(epochs, train_acc, marker='o', label=f'Train Acc ({aug_type})')
    plt.plot(epochs, val_acc, marker='x', linestyle='--', label=f'Val Acc ({aug_type})')
    
    best_val_acc = max(val_acc)
    print(f"The best validation accuracy for {aug_type} is: {best_val_acc:.2%}")
    
plt.title('Training and Validation Accuracy Across Augmentations')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
# Plotting variables
epochs = list(range(1, 11))  # Assuming there are 10 epochs
lr_fc = data['lr_fc']
schedulers = data['scheduler_type']
final_acc = data['final_acc']

# Determine unique learning rate factors
unique_lr_fc = sorted(set(lr_fc))

# Loop through each unique learning rate factor
for lr in unique_lr_fc:
    plt.figure(figsize=(10, 6))
    for i, (lrfc, sched) in enumerate(zip(lr_fc, schedulers)):
        if lrfc == lr:
            train_loss = final_acc[i]['train_loss']
            val_loss = final_acc[i]['val_loss']
            plt.plot(epochs, train_loss, marker='o', label=f'Train Loss - {sched}')
            plt.plot(epochs, val_loss, marker='x', linestyle='--', label=f'Val Loss - {sched}')
    plt.title(f'Training and Validation Loss for lr_fc = {lr}')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()

    plt.figure(figsize=(10, 6))
    for i, (lrfc, sched) in enumerate(zip(lr_fc, schedulers)):
        if lrfc == lr:
            train_acc = final_acc[i]['train_acc']
            val_acc = final_acc[i]['val_acc']
            plt.plot(epochs, train_acc,marker='o', label=f'Train Acc - {sched}')
            plt.plot(epochs, val_acc, marker='x', linestyle='--', label=f'Val Acc - {sched}')
    plt.title(f'Training and Validation Accuracy for lr_fc = {lr}')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()
```

```python
lr_fc = data['lr_fc']
schedulers = data['scheduler_type']

# Assuming the data structure in 'final_acc' is the same
epochs = list(range(1, 11))  # Assuming there are 10 epochs

# Organize data by learning rate and scheduler type
unique_lr_fc = sorted(set(lr_fc))
unique_schedulers = sorted(set(schedulers))

# Create figures for each lr_fc with subplots for each scheduler type
for lr in unique_lr_fc:
    fig, axs = plt.subplots(2, len(unique_schedulers), figsize=(20, 10), sharex=True, sharey='row')

    for j, scheduler in enumerate(unique_schedulers):
        # Filter data for current lr_fc and scheduler
        indices = [i for i, (lrfc, sch) in enumerate(zip(lr_fc, schedulers)) if lrfc == lr and sch == scheduler]
        for index in indices:
            # Plot training and validation accuracy and loss
            train_loss = data['final_acc'][index]['train_loss']
            val_loss = data['final_acc'][index]['val_loss']
            train_acc = data['final_acc'][index]['train_acc']
            val_acc = data['final_acc'][index]['val_acc']

            # Plotting
            axs[0, j].plot(epochs, train_loss, marker='o', label='Train Loss')
            axs[0, j].plot(epochs, val_loss, marker='x', linestyle='--', label='Val Loss')
            axs[1, j].plot(epochs, train_acc, marker='o', label='Train Acc')
            axs[1, j].plot(epochs, val_acc, marker='x', linestyle='--', label='Val Acc')

            # Titles and labels
            axs[0, j].set_title(f'{scheduler} Scheduler')
            axs[1, j].set_xlabel('Epoch')
            axs[0, 0].set_ylabel('Loss')
            axs[1, 0].set_ylabel('Accuracy')

            # Legend
            if index == indices[-1]:  # Add legend only to the last plot to avoid repetition
                axs[0, j].legend()
                axs[1, j].legend()

            # Find the best validation accuracy and print it
            best_val_acc = max(val_acc)
            print(f"Best validation accuracy for lr_fc={lr} with {scheduler} scheduler: {best_val_acc:.2%}")

    plt.suptitle(f'Plots for lr_fc = {lr}')
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
```
